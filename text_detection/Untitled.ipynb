```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test1.png')
```

```python
W = 2000
height, width, depth = img.shape
imgScale = W/width
newX,newY = img.shape[1]*imgScale, img.shape[0]*imgScale
```

```python
resize_img = cv2.resize(img, (int(newX), int(newY)))
```

```python
plt.imshow(resize_img)
```

```python
cv2.imwrite('img_resize.jpg', resize_img)
```

```python
img = cv2.imread('img_resize.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
_,img_binary = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(img_binary)
```

```python
median = cv2.medianBlur(img_binary,5)
plt.imshow(median)
```

```python
equalized_img = cv2.equalizeHist(median)
plt.imshow(equalized_img)
```

```python
cv2.imwrite("img.jpg", equalized_img)
```

```python
import numpy as np
img1 = np.bitwise_and(equalized_img,img)
```

```python
plt.imshow(img1)
```
